Créame un código en python de sistema de taxi siguiendo la historia de usuario anterior y sus criterios







```python
import random
from math import radians, sin, cos, sqrt, atan2

class Taxi:
    def __init__(self, driver_name, taxi_number, rating, location):
        self.driver_name = driver_name
        self.taxi_number = taxi_number
        self.rating = rating
        self.location = location

class Passenger:
    def __init__(self, name, current_location, destination):
        self.name = name
        self.current_location = current_location
        self.destination = destination

class TaxiSystem:
    def __init__(self):
        self.taxis = []
        self.passengers = []

    def request_taxi(self, passenger):
        available_taxis = [taxi for taxi in self.taxis if taxi.rating >= 3.5]
        if available_taxis:
            selected_taxi = random.choice(available_taxis)
            print(f"Taxi {selected_taxi.taxi_number} driven by {selected_taxi.driver_name} is on its way to pick you up.")
            self.passengers.append(passenger)
            # Calculate estimated time of arrival
            eta = self.calculate_distance(selected_taxi.location, passenger.current_location) / 60  # Assuming speed of 60 km/h
            print(f"Estimated time of arrival: {eta:.2f} minutes")
        else:
            print("Sorry, no taxis are available at the moment.")

    def calculate_distance(self, loc1, loc2):
        # Approximate radius of the Earth in km
        R = 6371.0

        lat1 = radians(loc1[0])
        lon1 = radians(loc1[1])
        lat2 = radians(loc2[0])
        lon2 = radians(loc2[1])

        dlon = lon2 - lon1
        dlat = lat2 - lat1

        a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
        c = 2 * atan2(sqrt(a), sqrt(1 - a))

        distance = R * c
        return distance

    def display_available_taxis(self, passenger_location):
        print("Available Taxis:")
        for taxi in self.taxis:
            distance = self.calculate_distance(taxi.location, passenger_location)
            print(f"Taxi {taxi.taxi_number} - {distance:.2f} km away")

    def assign_driver(self, taxi, passenger):
        print(f"Driver {taxi.driver_name} assigned to passenger {passenger.name}.")

    def rate_driver(self, passenger, taxi, rating, comment):
        taxi.rating = (taxi.rating + rating) / 2
        print(f"Thank you for your feedback, {passenger.name}! Your rating has been submitted.")

# Test the system
system = TaxiSystem()
system.taxis.append(Taxi("John", "TX123", 4.5, (40.7128, -74.0060)))  # New York City coordinates
system.taxis.append(Taxi("Alice", "TX456", 3.8, (34.0522, -118.2437)))  # Los Angeles coordinates

passenger = Passenger("Emma", (40.730610, -73.935242), (40.7580, -73.9855))  # New York to Manhattan coordinates

print("Passenger requests a taxi:")
system.request_taxi(passenger)

print("\nPassenger checks available taxis:")
system.display_available_taxis(passenger.current_location)

print("\nAssigning driver to passenger:")
system.assign_driver(system.taxis[0], passenger)

print("\nPassenger rates the driver:")
system.rate_driver(passenger, system.taxis[0], 5, "Great service!")


```
